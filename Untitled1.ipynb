```python
df_train
df_test

non_ordinal = ['Kelas Pekerja', 'Status Perkawinan', 'Pekerjaan', 'Jenis Kelamin' ]
ordinal = {'<=7jt':0,'>7jt':1}

df_train.replace('?',np.nan,inplace=True)
df_train['Pekerjaan'].fillna('Perbaikan Kerajinan')
df_train['Kelas Pekerja'].fillna('Wiraswasta')
df_train['Umur'].fillna(method='bfill')
df_train['Berat Akhir'].fillna(method='bfill')
df_train['Jmlh Tahun Pendidikan'].fillna(method='bfill')
df_train['Keuntungan Kapital'].fillna(method='bfill')
df_train['Kerugian Capital'].fillna(method='bfill')
df_train['Jam per Minggu'].fillna(method='bfill')

X_train = df_train.drop(['id','Gaji','Pendidikan'], axis=1)
Y_train = df_train['Gaji'].replace(ordinal)

#preprocess
X_train = pd.get_dummies(X_train, columns=non_ordinal)
X_train.head()
```

```python
df_train['Berat Akhir'].value_counts()
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV, RandomizedSearchCV
from sklearn.neighbors import KNeighborsClassifier
```

```python
model_knn = KNeighborsClassifier()
params = {'n_neighbors':np.arange(5,100), 'weights':['uniform','distance'], 'algorithm':['ball_tree','kd_tree','brute']}
gscv = GridSearchCV(model_knn, param_grid=params, scoring='roc_auc', cv=5)
gscv.fit(X_train,Y_train)
```

```python
print (gscv.best_score_)
print (gscv.best_params_)
print (gscv.best_estimator_)
```

```python
non_ordinal = ['Kelas Pekerja', 'Status Perkawinan', 'Pekerjaan', 'Jenis Kelamin' ]
ordinal = {'<=7jt':0,'>7jt':1}

df_test.replace('?',np.nan,inplace=True)
df_test['Pekerjaan'].fillna('Perbaikan Kerajinan')
df_test['Kelas Pekerja'].fillna('Wiraswasta')
df_test['Umur'].fillna(method='bfill')
df_test['Berat Akhir'].fillna(method='bfill')
df_test['Jmlh Tahun Pendidikan'].fillna(method='bfill')
df_test['Keuntungan Kapital'].fillna(method='bfill')
df_test['Kerugian Capital'].fillna(method='bfill')
df_test['Jam per Minggu'].fillna(method='bfill')
X_test = df_test.drop(['id','Pendidikan'], axis=1)

#preprocess
X_test = pd.get_dummies(X_test, columns=non_ordinal)
df_test.head()
```

```python
Y_test = gscv.predict(X_test)
```
