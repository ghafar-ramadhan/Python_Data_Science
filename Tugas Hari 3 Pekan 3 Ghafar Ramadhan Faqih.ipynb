```python
import matplotlib.pyplot as plt
import numpy as np
```

<h2>Soal 1: Multiple Subplots</h2>

Dengan data di bawah ini buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 200)
tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)
```

```python
# code here
fig, ax = plt.subplots(ncols=3, figsize=(20,10), sharex=True, sharey=True)
ax[0].plot(x,tan, color="blue", label="tanx")
ax[1].plot(x,sin, color="red", label="sinx")
ax[2].plot(x,cos, color="green", label="cosx")
ax[0].set_title("Grafik tanx")
ax[1].set_title("Grafik sinx")
fig.legend(loc='upper center')

plt.show()
```

<h2>Soal 2: Nested Axis</h2>

Dengan data di bawah ini, buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
```

```python
# code here
fig, ax = plt.subplots(figsize=(15,8))#, sharex=True, sharey=True)
ax.plot(x,y, color="blue", label="tanx")#

ins1=ax.inset_axes([-0.1, -0.1, 0.3, 0.3])
ins2=ax.inset_axes([0.9, -0.1, 0.3, 0.3])
ins3=ax.inset_axes([0.9, 0.9, 0.3, 0.3])
ins4=ax.inset_axes([-0.1, 0.9, 0.3, 0.3])

ins1.plot(x,y2, "r", x, y, "b")
ins2.plot(x,y3, "g", x, y, "b")
ins3.plot(x,y4, "k", x, y, "b")
ins4.plot(x,y5, "y", x, y, "b")

ins1.set_title("Grafik CosX^2")
ins2.set_title("Grafik CosX^3")
ins3.set_title("Grafik CosX^4")
ins4.set_title("Grafik CosX^5")
ax.set_title("Grafik CosX")
plt.show()
```
